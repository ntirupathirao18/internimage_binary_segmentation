```python
import pandas as pd
import shutil
import os
def filter_data(csv_path, input_path, type_run) :
  df = pd.read_csv(csv_path)
  imageids_list = df['new_filename'].tolist()
  folders_names = ['image','mask']
  for indfolder in folders_names :
    if not os.path.exists(os.path.join(input_path,type_run, indfolder)) :
      os.makedirs(os.path.join(input_path, type_run,indfolder))
  for indid in imageids_list :
    shutil.copy(os.path.join(input_path, 'png_images', indid), os.path.join(input_path,type_run, 'image', indid))
    shutil.copy(os.path.join(input_path, 'png_masks', indid), os.path.join(input_path,type_run, 'mask', indid))

filter_data(r'/content/data/siim-acr-pneumothorax/stage_1_train_images.csv', r'/content/data/siim-acr-pneumothorax' , type_run = 'train')
filter_data(r'/content/data/siim-acr-pneumothorax/stage_1_test_images.csv', r'/content/data/siim-acr-pneumothorax' ,  type_run = 'test')
```

```python
from mmseg.apis import init_model, inference_model, show_result_pyplot

config_path = 'configs/pspnet/pspnet_r50-d8_4xb2-40k_cityscapes-512x1024.py'
checkpoint_path = 'checkpoints/pspnet_r50-d8_512x1024_40k_cityscapes_20200605_003338-2966598c.pth'
img_path = 'demo/demo.png'


# build the model from a config file and a checkpoint file
model = init_model(config_path, checkpoint_path, device='cuda:0')

# inference on given image
result = inference_model(model, img_path)

# display the segmentation result
vis_image = show_result_pyplot(model, img_path, result)
```
