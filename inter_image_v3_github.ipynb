```python
import pandas as pd
import shutil
import os
def filter_data(csv_path, input_path, type_run) :
  df = pd.read_csv(csv_path)
  imageids_list = df['new_filename'].tolist()
  folders_names = ['image','mask']
  for indfolder in folders_names :
    if not os.path.exists(os.path.join(input_path,type_run, indfolder)) :
      os.makedirs(os.path.join(input_path, type_run,indfolder))
  for indid in imageids_list :
    shutil.copy(os.path.join(input_path, 'png_images', indid), os.path.join(input_path,type_run, 'image', indid))
    shutil.copy(os.path.join(input_path, 'png_masks', indid), os.path.join(input_path,type_run, 'mask', indid))

filter_data(r'/content/data/siim-acr-pneumothorax/stage_1_train_images.csv', r'/content/data/siim-acr-pneumothorax' , type_run = 'train')
filter_data(r'/content/data/siim-acr-pneumothorax/stage_1_test_images.csv', r'/content/data/siim-acr-pneumothorax' ,  type_run = 'test')
```

```python
import pandas as pd
import shutil
import os
import cv2
from matplotlib import pyplot as plt
import numpy as np
def filter_data_v2(csv_path, input_path, type_run) :
  df = pd.read_csv(csv_path)
  imageids_list = df['new_filename'].tolist()
  folders_names = ['image','mask']
  for indfolder in folders_names :
    if not os.path.exists(os.path.join(input_path,type_run, indfolder)) :
      os.makedirs(os.path.join(input_path, type_run,indfolder))
  for indid in imageids_list :
    shutil.copy(os.path.join(input_path, 'png_images', indid), os.path.join(input_path,type_run, 'image', indid))
    
    mask = cv2.imread(os.path.join(input_path, 'png_masks', indid))
    b_mask = np.copy(mask)
    #d = (('[255,255,255]':[1,1,1]))
    #for old, new in d.items():
    b_mask[mask == [255,255,255]] = 200
    # b_mask[mask == [0, 0, 0]] = 0
    b_mask = cv2.cvtColor(b_mask, cv2.COLOR_BGR2GRAY)

    cv2.imwrite(os.path.join(input_path,type_run, 'mask', indid) , b_mask)
     

filter_data_v2(r'/content/data/siim-acr-pneumothorax/stage_1_train_images.csv', r'/content/data/siim-acr-pneumothorax' , type_run = 'train_v2')
filter_data_v2(r'/content/data/siim-acr-pneumothorax/stage_1_test_images.csv', r'/content/data/siim-acr-pneumothorax' ,  type_run = 'test_v2')
```
