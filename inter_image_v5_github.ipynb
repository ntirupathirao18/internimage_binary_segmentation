```python
import pandas as pd
import shutil
import os
import cv2
from matplotlib import pyplot as plt
import numpy as np
import glob 
def filter_mask_v2(input_path ) :
  path_lists = glob.glob(input_path+'/*.*')
  if not os.path.exists(os.path.join(r'/content/data/data/data','masks_class_v3')) :
      os.makedirs(os.path.join(r'/content/data/data/data','masks_class_v3'))
  for indpath in path_lists :
    mask = cv2.imread(indpath)
    b_mask = np.copy(mask)
    # #d = (('[255,255,255]':[1,1,1]))
    # #for old, new in d.items():
    b_mask[mask == [128,128,128]] = 1
    # # b_mask[mask == [0, 0, 0]] = 0
    # b_mask = cv2.cvtColor(b_mask, cv2.COLOR_BGR2GRAY)
    
    cv2.imwrite(indpath.replace('masks','masks_class_v3') , b_mask )
     

filter_mask_v2(r'/content/data/data/data/masks')
```

```python
import mmcv
import os.path as osp
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as mpatche
img_dir = 'images'
ann_dir = 'masks'
classes = ('background', 'finding')
palette = [[0, 0, 0], [128, 128, 128]]
data_root = '/content/data/data/data'


for file in mmcv.scandir(osp.join(data_root, ann_dir), suffix='.png'):
    seg_map = np.array(Image.open(osp.join(data_root, ann_dir, file)).convert('L'))
    seg_img = Image.fromarray(seg_map).convert('P')
    seg_img.putpalette(np.array(palette, dtype=np.uint8))
    seg_img.save(osp.join(data_root, ann_dir, file))
```

```python
from mmseg.apis import init_model, inference_model, show_result_pyplot

config_path = 'configs/pspnet/pspnet_r50-d8_4xb2-40k_cityscapes-512x1024.py'
checkpoint_path = 'checkpoints/pspnet_r50-d8_512x1024_40k_cityscapes_20200605_003338-2966598c.pth'
img_path = 'demo/demo.png'


# build the model from a config file and a checkpoint file
model = init_model(config_path, checkpoint_path, device='cuda:0')

# inference on given image
result = inference_model(model, img_path)

# display the segmentation result
vis_image = show_result_pyplot(model, img_path, result)
```
